```python
import os
import pandas as pd
import numpy as np

directory = "/content"
dfLines = pd.DataFrame(columns=("Lines","contains_number","nb_words","Relevant"))
lines = []
contains_number=[]
nb_words = []
contains_keyword = []
for filename in os.listdir(directory):
  if (filename.endswith(".txt")):
    with open(filename,'r',encoding='utf-8') as f:
        for line in f.readlines(): 
            if line.strip():
                lines.append(line.rstrip())
                contains_number.append(int(any(char.isdigit() for char in line)))
                nb_words.append(len(line.split()))
                    


        
dfLines["Lines"]=lines
dfLines["contains_number"]=contains_number
dfLines["nb_words"]=nb_words
#dfLines["Relevant"]=relevant
```

```python
techno = ["PYTHON", "R Studio", 'SCALA','C++','HADOOP','SPARK','JAVASCRIPT','DOCKER','HIVE','HBASE','MESOS','COUCHBASE', 'ELK','KUBERNETES','SQL','JAVA', 'J2EE', 'SPRING', 'SPRING BOOT', 'REST','HIBERNATE', 'GROOVY', 'GRAILS', 'ANGULAR', 'IAAS', 'AWS', 'ELK','IAC','PACKER', 'ANSIBLE', 'SALTSTACK', 'PUPPET', 'CHEF','DOCKER','RANCHEROS','PYTHON','BASH','SHELL','JENKINS','GITLAB', 'TRAVIS', 'XL-DEPLOY','KUBERNETES','AWS']
techno = [x.lower() for x in techno]

if any(ele for ele in techno if(ele in str.lower())):
  print("True")
else:
  print("False") 
```

```python
def feature_extraction(df):
    date =["janvier","fevrier","mars","avril","mai","juin","juillet","aout","septembre","octobre","novembre","décembre"]
    keyword = ["lieu","poste","Consultant", "Salaire", "€", "Adresse", "Postale", "Ville", "Val", "Paris", "Montant", "Avenue", "Boulevard", "Rue", "Quai"]
    keyword = [x.lower() for x in keyword]
    entity = ["AIRBUS GROUP","AIR LIQUIDE","ALSTOM","ARCELORMITTAL","ATOS","AXA","BNP PARIBAS","BOUYGUES","CAPGEMINI","CARREFOUR","CREDIT AGRICOLE","DANONE","DASSAULT SYSTEMES","ENGIE","ESSILORLUXOTTICA","HERMES","KERING","L’OREAL","LEGRAND","LVMH","MICHELIN","ORANGE","PERNOD RICARD","PEUGEOT","PUBLICIS","RENAULT","SAFRAN","SAINT-GOBAIN","SANOFI","SCHNEIDER ELECTRIC","SOCIETE GENERALE","STMICROELECTRONICS","TELEPERFORMANCE","THALES","TOTAL","UNIBAIL-RODAMCO","VEOLIA","ENVIRONNEMENT","VINCI","VIVENDI","WORLDLINE"]
    entity = [x.lower() for x in entity]
    techno = ["PYTHON", "R Studio", 'SCALA','C++','HADOOP','SPARK','JAVASCRIPT','DOCKER','HIVE','HBASE','MESOS','COUCHBASE', 'ELK','KUBERNETES','SQL','JAVA', 'J2EE', 'SPRING', 'SPRING BOOT', 'REST','HIBERNATE', 'GROOVY', 'GRAILS', 'ANGULAR', 'IAAS', 'AWS', 'ELK','IAC','PACKER', 'ANSIBLE', 'SALTSTACK', 'PUPPET', 'CHEF','DOCKER','RANCHEROS','PYTHON','BASH','SHELL','JENKINS','GITLAB', 'TRAVIS', 'XL-DEPLOY','KUBERNETES','AWS']
    techno = [x.lower() for x in techno]
    special_characters = '!@#$%^&*()-+?_=,<>/"'
    for i in range(len(df)):
        ## Contains date
        str = df.loc[i,"Lines"]
        if any(ele for ele in date if(ele in str.lower())):
            df.loc[i,"contains_date"]=np.int64(1)
            df.loc[i,"Relevant"]=np.int64(1)
        else:
            df.loc[i,"contains_date"]=np.int64(0) 
        ## Start with small letter
        if str.strip()[0].isupper():
            df.loc[i,"starts_with_small_letter"]=np.int64(0)
        else:
            df.loc[i,"starts_with_small_letter"]=np.int64(1)
        ## Start with special char
        if any(c in special_characters for c in str):
            df.loc[i,"contains_special_char"]=np.int64(1)
        else:
            df.loc[i,"contains_special_char"]=np.int64(0)
        ## Start with number
        if str[0].isdigit():
            df.loc[i,"starts_with_number"]=np.int64(1)
        else:
            df.loc[i,"starts_with_number"]=np.int64(0)
        ## Technologies citées 
        if any(ele for ele in techno if(ele in str.lower())):
          df.loc[i,"contains_techno"]=np.int64(1)
          df.loc[i,"Relevant"]=np.int64(1)
        else:
          df.loc[i,"contains_techno"]=np.int64(0) 
        ## Contient un mot clé 
        if any(ele for ele in keyword if(ele in str.lower())):
          df.loc[i,"contains_keyword"]=np.int64(1)
          df.loc[i,"Relevant"]=np.int64(1)
        else:
          df.loc[i,"contains_keyword"]=np.int64(0)
        ## Contains entity names
        str = df.loc[i,"Lines"]
        if any(ele for ele in entity if(ele in str.lower())):
            df.loc[i,"contains_entity"]=np.int64(1)
            df.loc[i,"Relevant"]=np.int64(1)
        else:
            df.loc[i,"contains_entity"]=np.int64(0) 
          
        
    
```

```python
feature_extraction(dfLines)    
```

```python
dfLines["contains_number"] = dfLines["contains_number"].astype(int)
dfLines["contains_date"] = dfLines["contains_date"].astype(int)
dfLines["starts_with_small_letter"] = dfLines["starts_with_small_letter"].astype(int)
dfLines["contains_special_char"] = dfLines["contains_special_char"].astype(int)
dfLines["starts_with_number"] = dfLines["starts_with_number"].astype(int)
dfLines["contains_techno"] = dfLines["contains_techno"].astype(int)
dfLines["contains_entity"] = dfLines["contains_entity"].astype(int)
dfLines["contains_keyword"] = dfLines["contains_keyword"].astype(int)
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

X=dfLines[['contains_number', 'nb_words', 'contains_date', 'starts_with_small_letter','contains_special_char','starts_with_number','contains_techno','contains_entity','contains_keyword']]  # Features
y=dfLines['Relevant']  # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5) 
```

```python
y_train= y_train.astype(int)
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
y_test= np.array(y_test)
y_test = y_test.astype(np.int64)
```

```python
y_test
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
